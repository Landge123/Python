Subarrays as no-copy views

```python
print(x2)
```

```python
x2_sub = x2[:2,:2]
print(x2_sub)
```

```python
x2_sub [0,0] = 99
x2_sub
```

```python
print(x2)
```

```python
x2_sub_copy = x2[:2,:2].copy()
print(x2_sub_copy)
```

```python
x2_sub_copy[0,0] = 42
print(x2_sub_copy)
```
